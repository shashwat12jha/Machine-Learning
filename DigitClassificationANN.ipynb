```python
import tensorflow
from tensorflow import keras ##jisme neural network inbuilt hota hai
from tensorflow.keras import Sequential
from tensorflow.keras.layers import Dense,Flatten ##to add nn
```

```python
(X_train,y_train),(X_test,y_test) = keras.datasets.mnist.load_data() ##very popular dataset
```

```python
X_test.shape
```

```python
import matplotlib.pyplot as plt
plt.imshow(X_train[101])
```

```python
X_train = X_train/255
X_test = X_test/255
```

```python
##feature scaling so that weights can be calculated fast
```

```python
X_train
```

```python
model = Sequential()

model.add(Flatten(input_shape=(28,28))) ## 2-d ARRAY ko 1-D mein flatten kar rhe
model.add(Dense(128,activation='relu'))## humesha use relu as 1st hidden layer
model.add(Dense(10,activation='softmax'))
```

```python
model.summary()
```

```python
model.compile(loss='sparse_categorical_crossentropy',optimizer='Adam')
```

```python
history = model.fit(X_train,y_train,epochs=25,validation_split=0.2)
```

```python
y_prob = model.predict(X_test)
```

```python
y_pred = y_prob.argmax(axis=1)
```

```python
y_pred
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(y_test,y_pred)
```

```python
plt.plot(history.history['loss'])
plt.plot(history.history['val_loss'])
```
