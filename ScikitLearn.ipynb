```python
import numpy as np
import pandas as pd
from sklearn.impute import SimpleImputer
```

```python
train=pd.read_csv(r"C:\Users\shash\Downloads\archive (2)\train (1).csv")
test=pd.read_csv(r"C:\Users\shash\Downloads\archive (2)\test (1).csv")
print(train.shape)
print(test.shape)
```

```python
train.head()
```

```python
xtrain=train.drop(columns="SalePrice")
ytrain=train["SalePrice"]
```

```python
xtrain.shape
```

```python
numvar=xtrain.select_dtypes(include=[int,float]).columns
```

```python
numvar
```

```python
xtrain[numvar].isnull().sum()
```

```python
imputermean=SimpleImputer(strategy='mean')
```

```python
imputermean.fit(xtrain[numvar])
```

```python
imputermean.statistics_
```

```python
imputermean.transform(xtrain[numvar])
```

```python
xtrain[numvar]=imputermean.transform(xtrain[numvar])
test[numvar]=imputermean.transform(test[numvar])
```

```python
xtrain[numvar].isnull().sum()
```

```python
test[numvar].isnull().sum()
```

```python
catvar=xtrain.select_dtypes(include = [object]).columns
```

```python
catvar
```

```python
imputemean2=SimpleImputer(strategy='most_frequent')
```

```python
imputemean2.fit(xtrain[catvar])
```
