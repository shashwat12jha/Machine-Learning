```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv('diabetes.csv')
```

```python
df.head()
```

```python
df.corr()['Outcome']
```

```python
X=df.iloc[:,:-1].values
y=df.iloc[:,-1].values
```

```python
from sklearn.preprocessing import StandardScaler
scaler=StandardScaler()
```

```python
X=scaler.fit_transform(X)
```

```python
X.shape
```

```python
from sklearn.model_selection import train_test_split
Xtrain,Xtest,ytrain,ytest=train_test_split(X,y,test_size=0.2,random_state=1)
```

```python
import tensorflow
from tensorflow import keras
from keras import Sequential
from keras.layers import Dense
```

```python
model = Sequential()
model.add(Dense(32,activation='relu',input_dim=8))
model.add(Dense(1,activation='sigmoid'))
model.compile(optimizer='Adam',loss='binary_crossentropy',metrics=['accuracy'], run_eagerly=True)
```

```python
model.fit(Xtrain,ytrain,batch_size=32,epochs=10,validation_data=(Xtest,ytest))
```

#1

```python
import kerastuner as kt
```

```python
from keras.src.losses import binary_crossentropy
from tensorflow import keras

def build_model(hp):
  model=Sequential()
  model.add(keras.Input(shape=(8,))) # Changed input_dim to Input layer
  model.add(Dense(32,activation='relu'))
  model.add(Dense(1,activation='sigmoid'))
  model.compile(optimizer=hp.Choice('optimizer',['adam','sgd','resprop','adadelta']),loss=binary_crossentropy,metrics=['accuracy'])
  return model
```

```python
tuner=kt.RandomSearch(build_model,objective='val_accuracy',max_trials=5)
```

```python
tuner.search(Xtrain,ytrain,epochs=5,validation_data=(Xtest,ytest))
```

```python
model.summary()
```

```python
tuner.get_best_hyperparameters()[0].values
```

```python
model=tuner.get_best_models(num_models=1)[0]
```

```python
model.summary()
```

```python
model.fit(Xtrain,ytrain,batch_size=32,epochs=100,validation_data=(Xtest,ytest))
```

```python
def build_model(hp):
  model=Sequential()
  # Fix 1: Corrected hp.Int arguments
  units=hp.Int('units', min_value=8, max_value=128, step=8)
  # Fix 2: Changed input_dim to Input layer, consistent with previous fix
  model.add(keras.Input(shape=(8,)))
  model.add(Dense(units=units,activation='relu'))
  model.add(Dense(1,activation='sigmoid'))
  model.compile(optimizer=keras.optimizers.Adam(hp.Choice('learning_rate',values=[1e-2,1e-3])),loss=binary_crossentropy,metrics=['accuracy'])
  return model
```

```python
tuner=kt.RandomSearch(build_model,objective='val_accuracy',max_trials=5,directory='mydir')
```

```python
tuner.search(Xtrain,ytrain,epochs=5,validation_data=(Xtest,ytest))
```

```python
tuner.get_best_hyperparameters()[0].values
```

```python
model=tuner.get_best_models(num_models=1)[0]
```

```python
model.fit(Xtrain,ytrain,batch_size=32,epochs=100,initial_epoch=6,validation_data=(Xtest,ytest))
```

```python
def build_model(hp):
  model = Sequential()
  model.add(keras.Input(shape=(8,))) # Consistent with previous changes
  model.add(Dense(72,activation='relu'))
  for i in range(hp.Int('num_layers',min_value=1,max_value=10)):
    model.add(Dense(72,activation='relu'))
  model.add(Dense(1,activation='sigmoid'))
  model.compile(optimizer='rmsprop',loss='binary_crossentropy',metrics=['accuracy'])
  return model
```

```python
tuner=kt.RandomSearch(build_model,objective='val_accuracy',max_trials=5,directory='mydir',project_name='num_layers', overwrite=True)
```

```python
tuner.search(Xtrain,ytrain,epochs=5,validation_data=(Xtest,ytest))
```

```python
tuner.get_best_hyperparameters()[0].values
```

```python
model=tuner.get_best_models(num_models=1)[0]
```

```python
model.fit(Xtrain,ytrain,epochs=100,validation_data=(Xtest,ytest))
```
