```python
df=pd.read_csv('DailyDelhiClimate.csv')
```

```python
df.head()
```

```python
df.drop(columns=['humidity','wind_speed','meanpressure'],inplace=True)
```

```python
df
```

```python
df=df[['date','meantemp']]
df
```

```python
plt.scatter(df['date'],df['meantemp'],color='red')
```

```python
x1=df['meantemp'].ewm(alpha=0.4).mean()
x1
```

```python
df['ewma']=x1
```

```python
df
```

```python
plt.scatter(df['date'],df['meantemp'],color='red')
plt.plot(df['date'],x1,color='black',label='alpha=0.8')
plt.legend
plt.show
```
