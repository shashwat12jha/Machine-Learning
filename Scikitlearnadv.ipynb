```python
import numpy as np
import pandas as pd
from sklearn.impute import SimpleImputer
from sklearn.compose import ColumnTransformer
from sklearn.pipeline import Pipeline
```

```python
train=pd.read_csv(r"C:\Users\shash\Downloads\archive (2)\train (1).csv")
test=pd.read_csv(r"C:\Users\shash\Downloads\archive (2)\test (1).csv")
print(train.shape)
print(test.shape)
```

```python
xtrain=train.drop(columns="SalePrice")
ytrain=train["SalePrice"]
xtest=test.copy()
```

Missing Value Imputation

```python
numvar=xtrain.select_dtypes(include=[int,float]).columns
isnullsum=xtrain.isnull().sum()
numvarmiss=[var for var in numvar if isnullsum[var]>0]
```

```python
numvarmiss
numvarmean=['LotFrontage', 'MasVnrArea']
numvarmedian=['GarageYrBlt']
```

```python
catvar=xtrain.select_dtypes(include=[object]).columns
isnullsum=xtrain.isnull().sum()
catvarmiss=[var for var in catvar if isnullsum[var]>0]
```

```python
catvarmiss
```

```python
catvarmode=['Alley',
 'MasVnrType',
 'BsmtQual',
 'BsmtCond',
 'BsmtExposure',
 'BsmtFinType1',
 'BsmtFinType2','Electrical',
 'FireplaceQu']
catvarconst=[
 'GarageType',
 'GarageFinish',
 'GarageQual',
 'GarageCond',
 'PoolQC',
 'Fence',
 'MiscFeature']
```

```python
meanimputer=Pipeline(steps=[("imputer",SimpleImputer(strategy="mean"))])
medianimputer=Pipeline(steps=[("imputer",SimpleImputer(strategy="median"))])
modeimputer=Pipeline(steps=[("imputer",SimpleImputer(strategy="most_frequent"))])
constimputer=Pipeline(steps=[("imputer",SimpleImputer(strategy="constant",fill_value="MissingValue"))])       
```

```python
preprocessor = ColumnTransformer(
    transformers=[
        ("num_mean", meanimputer, numvarmean),
        ("num_median", medianimputer, numvarmedian),
        ("cat_mode", modeimputer, catvarmode),
        ("cat_const", constimputer, catvarconst)
    ]
)
```

```python
preprocessor.fit(xtrain)
```

```python
X_train_processed = preprocessor.transform(xtrain)
X_train_processed
```
